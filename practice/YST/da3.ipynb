```python
import numpy as np
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# 示例：生成一个模拟数据集
# 这里使用 make_blobs 创建一个二维数据集，有两个簇
X, y = make_blobs(n_samples=1000, centers=3, random_state=42)

# 将数据集划分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 1. 训练高斯混合模型（GMM）
gmm = GaussianMixture(n_components=3, covariance_type='full', random_state=42)
gmm.fit(X_train)

# 2. 预测训练集和测试集的类别
train_labels = gmm.predict(X_train)
test_labels = gmm.predict(X_test)

# 3. 可视化训练集和测试集的聚类结果
plt.figure(figsize=(12, 6))

# 可视化训练集的聚类结果
plt.subplot(1, 2, 1)
plt.scatter(X_train[:, 0], X_train[:, 1], c=train_labels, cmap='viridis', marker='o')
plt.title("Training Set Clustering (GMM)")

# 可视化测试集的聚类结果
plt.subplot(1, 2, 2)
plt.scatter(X_test[:, 0], X_test[:, 1], c=test_labels, cmap='viridis', marker='x')
plt.title("Test Set Clustering (GMM)")

plt.show()

# 4. 打印GMM的参数
print("Means of the Gaussian components:\n", gmm.means_)
print("\nCovariances of the Gaussian components:\n", gmm.covariances_)
print("\nMixing coefficients (weights):\n", gmm.weights_)
```
