```python
import matplotlib.pyplot as plt
plt.rc('font',family='Microsoft YaHei')
import seaborn as sns
import pandas as pd
```

```python
df = pd.read_excel('问卷第9题.xlsx')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 假设 df 是你的 DataFrame
num_columns = len(df.columns)
num_rows = 2
num_cols = num_columns // num_rows

fig, axs = plt.subplots(num_rows, num_cols, figsize=(12, 8))

# 遍历 DataFrame 的每一列
for i, column_name in enumerate(df.columns):
    # 获取列数据
    column_data = df[column_name]

    # 计算子图位置
    row_index = i // num_cols
    col_index = i % num_cols

    # 绘制概率密度图
    sns.distplot(column_data, ax=axs[row_index, col_index], kde=True, hist=True, kde_kws={"shade": True}, hist_kws={'range': (0, 100)})
    axs[row_index, col_index].set_title(column_name)
    axs[row_index, col_index].set_ylabel('')

    # 设置x轴范围为0到100，间距为10
    axs[row_index, col_index].set_xlim(0, 100)
    axs[row_index, col_index].set_xticks(range(0, 101, 10))
    
    axs[row_index, col_index].set_xlabel('')
    axs[row_index, col_index].set_ylabel('')
    # 设置y轴范围为0到0.4
    axs[row_index, col_index].set_ylim(0, 0.05)
    axs[row_index, col_index].grid(False)

# 隐藏多余的子图
for i in range(num_columns, num_rows * num_cols):
    row_index = i // num_cols
    col_index = i % num_cols
    axs[row_index, col_index].axis('off')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# 假设 df 是你的 DataFrame
for column_name, column_data in df.items():
    # 计算平均值和标准差
    mean_value = np.mean(column_data)
    std_deviation = np.std(column_data)

    # 计算3σ范围
    lower_bound = mean_value -   2*std_deviation
    upper_bound = mean_value +   2*std_deviation

    print(f"列 '{column_name}' 的2σ范围:", lower_bound, "到", upper_bound,'  mean:',mean_value)
```
